```python
import re
import json
import matplotlib.pyplot as plt
import codecs
import numpy as np
import jieba

def load_nerDict(fname):
    nerDict, cnt = [], 0
    with open(fname, 'r', encoding='utf-8') as f:
        for line in f:
            if line.strip():
                cnt += 1
                # print(line)
                nerDict.append(line.strip())
            # if cnt == 2:
            #     break
    print(cnt)
    return nerDict

def load_train_data(train_file, ner_file):
    news_id = set()
    entities_dict = set()
    with open(train_file, 'r', encoding='utf-8') as f:
        for line in f:
            data = json.loads(line)
            entities = data['coreEntityEmotions']
            for a in entities:
                entities_dict.add(a['entity'])
            # print(data['coreEntityEmotions'])
            # print(len(data['coreEntityEmotions']))
    print("entities_dict:{}".format(len(entities_dict)))
    nerDict = load_nerDict(ner_file)
    cnt = 0
    ner_not_in_nerdict = []
    for i in entities_dict:
        if i in nerDict:
            cnt += 1
        else:
            ner_not_in_nerdict.append(i)

    print("number of entities in nerDict:{}".format(cnt))
    print("number of entities not in nerDict:{}".format(len(entities_dict) - cnt))
#     print(ner_not_in_nerdict)
    return nerDict, ner_not_in_nerdict
```

```python
basedir = '../input_data'
train_file = basedir + '/coreEntityEmotion_train.txt' 
test_file = basedir + '/coreEntityEmotion_test_stage1.txt'
ner_file = basedir + '/nerDict.txt'
stop_words_file = basedir + '/stop_words_mine.txt'

nerDict, ner_not_in_nerdict = load_train_data(train_file, ner_file)
print(len(nerDict))
print(len(ner_not_in_nerdict))
```

```python
len_in_ner_not_in_nerdict = [len(i) for i in ner_not_in_nerdict]
```
